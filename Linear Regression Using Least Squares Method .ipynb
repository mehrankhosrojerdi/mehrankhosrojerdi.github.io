In this section we are dealing with the linear regression using Least Squares method. 
In fact we have a scatter plot hat we are looking for the best line that is fit with this scatter plot.

```python
import numpy as np 
import matplotlib.pyplot as plt

x = np.array([1,2,3,4,5,6,7])
y = np.array([4,6,7,6,7,6,9])

plt.scatter(x,y) 
plt.show()
```

For find the best line we need to find weight $w$ and bias $b$ in below form 
$$y = w\times x + b$$
which 
$$ w = \frac{n \sum xy -\sum x \sum y}{n \sum x^{2} - (\sum x)^2} $$
$$ b = \frac {\sum y - w \sum x}{n} $$
Therfore it is clear that we need to calculate $\sum x$ , $\sum y$, $\sum xy$, and $\sum x^2$

```python
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xy = np.dot(x,y)
sum_xx = np.dot(x,x)
n = x.shape[0]
```

Mow we can make $w$ and $b$

```python
w = ((n*sum_xy) - (sum_x*sum_y))/((n*sum_xx) - (sum_x**2))
b = (sum_y -( w*sum_x))/n
```

Let's have a look on result

```python
x = np.array([1,2,3,4,5,6,7])
y = np.array([4,6,7,6,7,6,9])
y_pred = w*x+b

plt.scatter(x,y) 
plt.plot(x,y_pred,'r')
plt.show()
```

Let's have a look on least square method with another example. Here we want to use Linear Regression as a linear model in sklearn library. The dataset that we are using here is diabetes dataset that is prepared by sklearn. 

```python
import sklearn.datasets
import sklearn.linear_model
import sklearn.metrics
import matplotlib.pyplot as plt
import sklearn.preprocessing
import sklearn.model_selection
import numpy as np

diabetes = sklearn.datasets.load_diabetes()

X = diabetes.data[:,4] # chose only a feature
X = X.reshape(-1, 1) 
y = diabetes.target

X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, train_size=0.8, random_state=42)

regr = sklearn.linear_model.LinearRegression()
regr.fit(X_train, y_train)

y_pred = regr.predict(X_test)

acc = sklearn.metrics.r2_score(y_pred = y_pred, y_true = y_test)

print(f'the accuracy of this model is {acc}')
plt.scatter(X_test, y_test,)
plt.plot(X_test, y_pred, 'r')
plt.show()
```
