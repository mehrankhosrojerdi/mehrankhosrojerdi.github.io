In this section we are dealing with the linear regression using Least Squares method. 
In fact we have a scatter plot hat we are looking for the best line that is fit with this scatter plot.

```python
import numpy as np 
import matplotlib.pyplot as plt

x = np.array([1,2,3,4,5,6,7])
y = np.array([4,6,7,6,7,6,9])

plt.scatter(x,y) 
plt.show()
```

For find the best line we need to find weight $w$ and bias $b$ in below form 
$$y = w\times x + b$$
which 
$$ w = \frac{n \sum xy -\sum x \sum y}{n \sum x^{2} - (\sum x)^2} $$
$$ b = \frac {\sum y - w \sum x}{n} $$
Therfore it is clear that we need to calculate $\sum x$ , $\sum y$, $\sum xy$, and $\sum x^2$

```python
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xy = np.dot(x,y)
sum_xx = np.dot(x,x)
n = x.shape[0]
```

Mow we can make $w$ and $b$

```python
w = ((n*sum_xy) - (sum_x*sum_y))/((n*sum_xx) - (sum_x**2))
b = (sum_y -( w*sum_x))/n
```

Let's have a look on result

```python
x = np.array([1,2,3,4,5,6,7])
y = np.array([4,6,7,6,7,6,9])
y_pred = w*x+b

plt.scatter(x,y) 
plt.plot(x,y_pred,'r')
plt.show()
```
